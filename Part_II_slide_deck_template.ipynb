## Start Station Location


```python
df.plot('start_station_latitude','start_station_longitude', xlabel = 'Laltitude',ylabel = 'Longitude' , title = 'Start Station Location', kind = 'scatter' , c= 'green');
```

## Bike share per age

## Trip Duration per Age For each gender



```python
sb.lineplot(df.age,df.duration_minute , hue = df.member_gender,ci = False).set(title ='Trip Duration per Age For each gender' ,xlabel ='Member Age' ,ylabel= 'Duration(Minute)');
```
